```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/mymoviedb_clean.csv")
```

```python
df.head()
```

```python
# Explore counts by type (movie/TV), year trends and top genres.
```

```python
# Counts by type (Movie/TV)
df['Genre'].value_counts()
```

```python
# Year Trend
Year_Trend = df.groupby('Release_Date')['Popularity'].mean().reset_index()
```

```python
Year_Trend
```

```python
Year_Trend['Popularity'] = Year_Trend['Popularity'].round().astype(int)
```

```python
Year_Trend
```

```python

plt.figure(figsize=(10, 6))
plt.plot(Year_Trend['Release_Date'], Year_Trend['Popularity'])
plt.title('Year Trend')
plt.xlabel('Year')
plt.show()
```

```python
# Top 5 Genre
df['Genre'].value_counts()[:5]
```

```python
# Visualize content growth over time and runtime distributions.
```

```python
movies_per_year = df['Release_Date'].value_counts().reset_index()
movies_per_year.columns = ['Year', 'Number of Movies']
movies_per_year = movies_per_year.sort_values(by='Year').reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=movies_per_year['Year'], y=movies_per_year['Number of Movies'], data=movies_per_year)
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```
